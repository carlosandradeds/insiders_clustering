```python
import pandas as pd
import seaborn as sns
import re

from matplotlib            import pyplot as plt
```

```python
data = pd.read_csv('C:/Users/kaka/Documents/candidatos/consulta_cand_2022_BRASIL.csv', sep=';', encoding='unicode_escape' )
```

```python
data.columns
```

```python
data['DS_GENERO'].unique()
```

```python
aux = data[data['DS_GENERO'] != 'NÃO DIVULGÁVEL' ]
```

```python
ax = sns.countplot( x='DS_GENERO', data=aux)

plt.title('Distribution of  Gender')
plt.xlabel('Number of Candidates for Gender')

total = len(data['DS_GENERO'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width()
        y = p.get_height()
        ax.annotate(percentage, (x, y))

plt.show();
```

```python
data['NM_PARTIDO'].unique()
```

```python
ax = sns.countplot( x='SG_PARTIDO', data=data)

sns.set(rc = {'figure.figsize':(15,8)})

plt.title('Distribuição dos Partidos')
plt.xlabel('Partidos')
plt.ylabel('N Candidatos')

total = len(data['SG_PARTIDO'])
for p in ax.patches:
        percentage = '{:.1f}%'.format(100 * p.get_height()/total)
        x = p.get_x() + p.get_width()
        y = p.get_height()
        ax.annotate(percentage, (x, y))

      
ax.set_xticklabels(ax.get_xticklabels(), rotation=90)
plt.figure(figsize = (15,8))
plt.show();
```
